```python
import seaborn as sns
sns.set()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

## Avg of pepole who still a live on the ship

```python
still_alive=titanic.groupby('alive',as_index=False)['age'].mean()
still_alive
```

```python
sns.barplot(x='alive',y='age',data=still_alive);
```

## Number of Male and Female who was alone

```python
sns.barplot(x='sex',y='alone',data=titanic);
```
